```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import MultipleLocator
from astropy.table import join, vstack, Table

from my_progs.stat_func.rms_calc import rms_calc
from my_progs.vlbi.ts_func import get_ts
from my_progs.catalog.read_icrf import read_icrf3
```

```python
icrf3sx = read_icrf3(wv="sx")

icrf3def = icrf3sx[icrf3sx["type"] == "D"]
```

```python
sou_ts = get_ts("sou-ti-concat", "/Users/Neo/Astronomy/data/vlbi/opa/ts-sou")

sou_ts_g = sou_ts.group_by("db_name")
```

Session-wise CRF orientation aqnd glide

```python
sess_orient_def = Table.read("../data/sess-wise-orient-def.txt", format="ascii")
sess_orient_def.sort("jyear")

sess_orient_all = Table.read("../data/sess-wise-orient-all.txt", format="ascii")
sess_orient_all.sort("jyear")
```

```python
sess_name = []
sess_epo = []
num_sou = []
num_def = []
```

```python
for group in sou_ts_g.groups:
    sess_name.append(group["db_name"][0])
    sess_epo.append(group["jyear"][0])
    
    num_sou.append(len(group))
    
    group_def = join(group, icrf3def, keys="iers_name")
    num_def.append(len(group_def))
```

```python
fig, ax = plt.subplots()

ax.plot(sess_epo, num_sou, "bo", ms=1, lw=0.5, label="All")
ax.plot(sess_epo, num_def, "rs", ms=1, lw=0.5, label="Defining")
ax.plot(sess_orient_def["jyear"], sess_orient_def["num_sou"], "gx", ms=1, lw=0.5, label="Used")

ax.set_yscale("log")

ax.set_xlabel("Epoch (year)", fontsize=12)
ax.set_ylabel("No. sources", fontsize=12)
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.legend()

plt.tight_layout()
plt.savefig("../plots/num_sou_in_sess.eps")
```

```python
number = np.arange(len(sess_orient_def))
binsize = 100

nb_bin = np.trunc(number / binsize)

tab_g = sess_orient_def.group_by(nb_bin)

N = len(tab_g.groups)

epoch_med = np.zeros(N)
rx_wrms = np.zeros(N)
rx_mean = np.zeros(N)
rx_std = np.zeros(N)
ry_wrms = np.zeros(N)
ry_mean = np.zeros(N)
ry_std = np.zeros(N)
rz_wrms = np.zeros(N)
rz_mean = np.zeros(N)
rz_std = np.zeros(N)
gx_wrms = np.zeros(N)
gx_mean = np.zeros(N)
gx_std = np.zeros(N)
gy_wrms = np.zeros(N)
gy_mean = np.zeros(N)
gy_std = np.zeros(N)
gz_wrms = np.zeros(N)
gz_mean = np.zeros(N)
gz_std = np.zeros(N)

for i, group in enumerate(tab_g.groups):

    epoch_med[i] = np.median(group["jyear"])

    rx_mean[i], rx_wrms[i], rx_std[i] = rms_calc(group["rx"], group["rx_err"])
    ry_mean[i], ry_wrms[i], ry_std[i] = rms_calc(group["ry"], group["ry_err"])
    rz_mean[i], rz_wrms[i], rz_std[i] = rms_calc(group["rz"], group["rz_err"])
    gx_mean[i], gx_wrms[i], gx_std[i] = rms_calc(group["gx"], group["gx_err"])
    gy_mean[i], gy_wrms[i], gy_std[i] = rms_calc(group["gy"], group["gy_err"])
    gz_mean[i], gz_wrms[i], gz_std[i] = rms_calc(group["gz"], group["gz_err"])
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(12, 12), nrows=3, sharex=True, sharey=True)

ax0.errorbar(sess_orient_def["jyear"], sess_orient_def["rx"], yerr=sess_orient_def["rx_err"], color="blue",
            ms=1, fmt="s", elinewidth=0.1, lw=0.2, label="$R_1$")
ax1.errorbar(sess_orient_def["jyear"], sess_orient_def["ry"], yerr=sess_orient_def["ry_err"], color="black",
            ms=1, fmt="o", elinewidth=0.1, lw=0.2, label="$R_2$")
ax2.errorbar(sess_orient_def["jyear"], sess_orient_def["rz"], yerr=sess_orient_def["rz_err"], color="g",
            ms=1, fmt="^", elinewidth=0.1, lw=0.2, label="$R_3$")

ax0.plot(epoch_med, rx_std, "r--", lw=5)
ax1.plot(epoch_med, ry_std, "r--", lw=5)
ax2.plot(epoch_med, rz_std, "r--", lw=5)

ax0.xaxis.set_minor_locator(MultipleLocator(1))
ax1.xaxis.set_minor_locator(MultipleLocator(1))
ax2.xaxis.set_minor_locator(MultipleLocator(1))

ax2.set_xlabel("Epoch (year)", fontsize=12)
ax0.set_ylabel("$\\epsilon_x$  ($\mathrm{\mu as}$)", fontsize=15)
ax1.set_ylabel("$\\epsilon_y$ ($\mathrm{\mu as}$)", fontsize=15)
ax2.set_ylabel("$\\epsilon_z$ ($\mathrm{\mu as}$)", fontsize=15)

ax0.axis([1985, 2022, -1000, 1000])

plt.tight_layout()
plt.savefig("../plots/orient-from-sess-crf.eps")
```
