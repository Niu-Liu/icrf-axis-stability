```python
import matplotlib.pyplot as plt
import numpy as np

from astropy.table import join, setdiff, Table
import astropy.units as u

# My progs
from myprogs.vlbi.read_sou import read_sou
from myprogs.catalog.read_icrf import read_icrf3
from myprogs.catalog.pos_diff import radio_cat_diff_calc
```

```python
icrf3sx = read_icrf3(wv="sx")

icrf3def = icrf3sx[icrf3sx["type"] == "D"]
```

```python
def read_table(year):
    glo_sou = Table.read("../data/yearly-ts/{:d}.dat".format(year), format="ascii")

    # Add unit information
    glo_sou["ra"].unit = u.deg
    glo_sou["dec"].unit = u.deg
    glo_sou["ra_err"].unit = u.mas
    glo_sou["dec_err"].unit = u.mas

    return glo_sou
```

```python
year = 2000
glo_sou = read_table(year)
pos_oft00 = radio_cat_diff_calc(glo_sou, icrf3def, sou_name="iers_name")

year = 2001
glo_sou = read_table(year)
pos_oft01 = radio_cat_diff_calc(glo_sou, icrf3def, sou_name="iers_name")

year = 2002
glo_sou = read_table(year)
pos_oft02 = radio_cat_diff_calc(glo_sou, icrf3def, sou_name="iers_name")

year = 2003
glo_sou = read_table(year)
pos_oft03 = radio_cat_diff_calc(glo_sou, icrf3def, sou_name="iers_name")

year = 2004
glo_sou = read_table(year)
pos_oft04 = radio_cat_diff_calc(glo_sou, icrf3def, sou_name="iers_name")

year = 2017
glo_sou = read_table(year)
pos_oft17 = radio_cat_diff_calc(glo_sou, icrf3def, sou_name="iers_name")

year = 2018
glo_sou = read_table(year)
pos_oft18 = radio_cat_diff_calc(glo_sou, icrf3def, sou_name="iers_name")

year = 2019
glo_sou = read_table(year)
pos_oft19 = radio_cat_diff_calc(glo_sou, icrf3def, sou_name="iers_name")
```

```python
# Difference in RA and decl.
year = 2000
fig, (ax0, ax1) = plt.subplots(figsize=(8, 6), nrows=2, sharex=True, sharey=True)

ax0.errorbar(pos_oft00['ra'], pos_oft00['dra'], yerr=pos_oft00['dra_err'],
             fmt='.', color='b', elinewidth=0.2, ms=2)
ax1.errorbar(pos_oft00['ra'], pos_oft00['ddec'], yerr=pos_oft00['ddec_err'],
             fmt='.', color='b', elinewidth=0.2, ms=2)

ax0.set_xticks(np.arange(0, 361, 60))

xmax = 10
ax0.set_ylim([-xmax, xmax])

ax1.set_xlabel('RA ($^\circ$)', fontsize=15)
ax0.set_ylabel('$\Delta\\alpha*$ (mas)', fontsize=15)
ax1.set_ylabel('$\Delta\\delta$ (mas)', fontsize=15)

ax0.set_title("{:d}-ts $-$ ICRF3 S/X".format(year), fontsize=15)
fig.tight_layout()

# Difference in RA and decl.
fig, (ax0, ax1) = plt.subplots(figsize=(8, 6), nrows=2, sharex=True, sharey=True)

ax0.errorbar(pos_oft00['dec'], pos_oft00['dra'], yerr=pos_oft00['dra_err'],
             fmt='.', color='b', elinewidth=0.2, ms=2)
ax1.errorbar(pos_oft00['dec'], pos_oft00['ddec'], yerr=pos_oft00['ddec_err'],
             fmt='.', color='b', elinewidth=0.2, ms=2)

ax0.set_xticks(np.arange(-90, 91, 60))
ax0.set_ylim([-xmax, xmax])

ax1.set_xlabel('Declination ($^\circ$)', fontsize=15)
ax0.set_ylabel('$\Delta\\alpha*$ (mas)', fontsize=15)
ax1.set_ylabel('$\Delta\\delta$ (mas)', fontsize=15)

# ax0.set_title("{:d}-c $-$ ICRF3 S/X".format(year+1), fontsize=15)
fig.tight_layout()
```
