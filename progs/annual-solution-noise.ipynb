I found that a VLBI global solution based on a small subset of data (e.g., with 2-year window) would show a formal error floor ~ 1 mas (OPA11 and OPA19).
It is hard to explain, and I will check whether similar cases happen to my annual solutions.

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np

# My modules
from my_progs.vlbi.read_sou import read_sou
```

```python
num_sou = []
mederr_ra = []
mederr_dec = []
mederr_pos = []
```

```python
year = 1985
opa = read_sou("../data/yearly/{:d}-y.sou".format(year), drop_few_obs=True)

fig, ax = plt.subplots()

ax.scatter(opa["used_ses"], opa["pos_err"], marker="o",
           s=3, lw=0.1)

ax.set_ylim([0.001, 100])
ax.set_xscale("log")
ax.set_yscale("log")

ax.set_xlabel("$N_\mathrm{ses}$", fontsize=12)
ax.set_ylabel("$\sigma_\mathrm{pos,max}$ (mas)", fontsize=12)
ax.set_title(year)

num_sou.append(len(opa))
mederr_ra.append(np.median(opa["ra_err"]))
mederr_dec.append(np.median(opa["dec_err"]))
mederr_pos.append(np.median(opa["pos_err"]))
```

```python
def get_sol_stat(rms_file):
    """
    """
    
    with open(rms_file, "r") as f_in:
        line = f_in.readlines()
        
        data = line[1].strip("\n").split()
        
        n_ses = len(line) - 3
        n_obs = np.int("".join(data[2:4]))
        tau_wrms = np.float(data[4])
        chi2 = np.float(data[8])
        
    return n_ses, n_obs, tau_wrms, chi2
```

```python
data_dir = "/Users/Neo/Astronomy/data/vlbi/vlbi2_server/yearly-crf/yearly/"
```

```python
n_ses = np.empty((epoch.size,), dtype=np.int)
n_obs = np.empty((epoch.size,), dtype=np.int)
tau_wrms = np.empty((epoch.size,), dtype=np.float)
chi2 = np.empty((epoch.size,), dtype=np.float)
```

```python
for i, year in enumerate(epoch):
    n_ses[i], n_obs[i], tau_wrms[i], chi2[i] = get_sol_stat(
        "{:s}/{:.0f}-y/{:.0f}-y.rms".format(data_dir, year, year))
```
