```python
import matplotlib.pyplot as plt
import numpy as np

from astropy.table import join, setdiff

# My progs
from my_progs.vlbi.read_sou import read_sou
from my_progs.catalog.read_icrf import read_icrf3
from my_progs.catalog.pos_diff import radio_cat_diff_calc
```

```python
icrf3sx = read_icrf3(wv="sx")

icrf3def = icrf3sx[icrf3sx["type"] == "D"]
```

```python
year = 2000
glo_sou00 = read_sou("../data/yearly-solution/{:d}-c.sou".format(year))
pos_oft00 = radio_cat_diff_calc(glo_sou00, icrf3def, sou_name="iers_name")

year = year + 1
glo_sou01 = read_sou("../data/yearly-solution/{:d}-c.sou".format(year))
pos_oft01 = radio_cat_diff_calc(glo_sou01, icrf3def, sou_name="iers_name")

year = year + 1
glo_sou02 = read_sou("../data/yearly-solution/{:d}-c.sou".format(year))
pos_oft02 = radio_cat_diff_calc(glo_sou02, icrf3def, sou_name="iers_name")

year = year + 1
glo_sou03 = read_sou("../data/yearly-solution/{:d}-c.sou".format(year))
pos_oft03 = radio_cat_diff_calc(glo_sou03, icrf3def, sou_name="iers_name")

# year = year + 1
# glo_sou04 = read_sou("../data/yearly-solution/{:d}-c.sou".format(year))
# # glo_sou03 = read_sou("../data/yearly-solution/{:d}-c.sou".format(year))
# pos_oft04 = radio_cat_diff_calc(glo_sou03, icrf3def, sou_name="iers_name")
```

```python
# Difference in RA and decl.
fig, (ax0, ax1) = plt.subplots(figsize=(10, 6), ncols=2, sharex=True, sharey=True)

lim = 2.5
x = np.arange(-lim, lim, 0.1)
ax0.plot(x, x, "r--", lw=0.5)
ax1.plot(x, x, "r--", lw=0.5)

ax0.errorbar(com_sou['dra_0'], com_sou['dra_1'],
             xerr=com_sou['dra_err_0'], yerr=com_sou['dra_err_1'],
             fmt='.', color='b', elinewidth=0.2, ms=2)
ax1.errorbar(com_sou['ddec_0'], com_sou['ddec_1'],
             xerr=com_sou['ddec_err_0'], yerr=com_sou['ddec_err_1'],
             fmt='.', color='b', elinewidth=0.2, ms=2)

ax0.axis("square")
ax1.axis("square")
ax1.axis([-lim, lim, -lim, lim])

ax0.set_xlabel('$\Delta\\alpha*$ (2001-c, mas)', fontsize=15)
ax0.set_ylabel('$\Delta\\alpha*$ (2002-c, mas)', fontsize=15)
ax1.set_xlabel('$\Delta\\delta$ (2001-c, mas)', fontsize=15)
ax1.set_ylabel('$\Delta\\delta$ (2002-c, mas)', fontsize=15)

fig.tight_layout()
```
