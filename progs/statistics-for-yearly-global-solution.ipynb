I made several plots in order to show the main statistics of the VLBI global solution evolves with the time.

These statistics include:

    - number of sessions
    - number of observations
    - postfit rms of delay
    - Chi per number of degrees of freedom
    - number of sources in catalogs
    - relationship of formal uncertainties versus number of sessions
    - median formal uncertainties

```python
import numpy as np

from astropy.table import Table
import seaborn as sns
```

```python
from my_progs.vlbi.rms_func import get_stats_from_rms
```

```python
sol_dir = "/Users/Neo/Astronomy/data/vlbi/vlbi2_server/yearly-crf/continued"
```

```python
year = np.arange(1984, 2021)

num_sess = np.zeros_like(year)
num_obs = np.zeros_like(year)
rms = np.zeros_like(year, dtype=float)
chi = np.zeros_like(year, dtype=float)
```

```python
for i, yeari in enumerate(year):
    rms_file = "{}/{}-c/{}-c.rms".format(sol_dir, yeari, yeari)
    stats = get_stats_from_rms(rms_file)

    num_sess[i] = stats["num_sess"]
    num_obs[i] = stats["num_obs"]
    rms[i] = stats["postfit_rms"]
    chi[i] = stats["chi_ndg"]
```

```python
sol_stats = Table(
    [year, num_sess, num_obs, rms, chi],
    names=["year", "num_sess", "num_obs", "postfit_rms", "chi_ndg"])
```

```python
sol_stats.write("../logs/yearly-sol-stats.dat", format="ascii", overwrite=True)
```

```python
sol_stats = sol_stats.to_pandas()
```

```python
sns.relplot(x="year", y="num_sess", data=sol_stats, color="b")
```

```python
sns.relplot(x="year", y="num_obs", data=sol_stats, color="b")
```

```python
sns.relplot(x="year", y="postfit_rms", data=sol_stats, color="b")
```

```python
sns.relplot(x="year", y="chi_ndg", data=sol_stats, color="b")
```

```python
import matplotlib.pyplot as plt

# My modules
from my_progs.vlbi.sou_func import read_sou
```

```python
num_sou = []
mederr_ra = []
mederr_dec = []
mederr_pos = []
```

```python
def simple_plot(year, num_sou, mederr_ra, mederr_dec, mederr_pos):
    """
    """

    opa = read_sou("../data/yearly-solution/{:d}-c.sou".format(year),
                   drop_few_obs=True)

    fig, ax = plt.subplots()

    ax.scatter(opa["used_ses"], opa["pos_err"], marker="o", s=3, color="b")

    ax.set_ylim([0.005, 100])
    ax.set_xscale("log")
    ax.set_yscale("log")

    ax.set_xlabel("$N_\mathrm{ses}$", fontsize=12)
    ax.set_ylabel("$\sigma_\mathrm{pos,max}$ (mas)", fontsize=12)
    ax.set_title("1979.08 - {}.01".format(year + 1), fontsize=12)

    plt.tight_layout()

    # Some statistics
    num_sou.append(len(opa))
    mederr_ra.append(np.median(opa["ra_err"]))
    mederr_dec.append(np.median(opa["dec_err"]))
    mederr_pos.append(np.median(opa["pos_err"]))

    return num_sou, mederr_ra, mederr_dec, mederr_pos
```
