```python
import matplotlib.pyplot as plt
# from matplotlib.ticker import MultipleLocator
import numpy as np
import statsmodels.api as sm
from astropy.table import Table, join

from statsmodels.iolib.table import (SimpleTable, default_txt_fmt)
import seaborn as sns

from my_progs.stat_func.rms_calc import rms_calc
```

```python
apm_tab = Table.read("../logs/ts_pm_fit_3sigma.dat", format="ascii.csv")

# convert mas/yr into muas/yr
apm_tab["pmra"] = apm_tab["pmra"] * 1e3
apm_tab["pmra_err"] = apm_tab["pmra_err"] * 1e3
apm_tab["pmdec"] = apm_tab["pmdec"] * 1e3
apm_tab["pmdec_err"] = apm_tab["pmdec_err"] * 1e3

mask = apm_tab["num_cln"] >= 5
apm_tab = apm_tab[mask]
```

```python
apm_tab1 = Table.read("../logs/ts_glo_pm_fit_3sigma.dat", format="ascii.csv")

# convert mas/yr into muas/yr
apm_tab1["pmra"] = apm_tab1["pmra"] * 1e3
apm_tab1["pmra_err"] = apm_tab1["pmra_err"] * 1e3
apm_tab1["pmdec"] = apm_tab1["pmdec"] * 1e3
apm_tab1["pmdec_err"] = apm_tab1["pmdec_err"] * 1e3


mask = apm_tab1["num_cln"] >= 5
apm_tab1 = apm_tab1[mask]
```

```python
apm_com_tab = join(apm_tab, apm_tab1, keys="iers_name")
```

```python
fig, (ax0, ax1) = plt.subplots(figsize=(12, 6), ncols=2)

lim = 100
ax0.axis("square")
ax0.axis([-lim, lim, -lim, lim])

ax1.axis("square")
ax1.axis([-lim, lim, -lim, lim])

x = np.arange(-lim, lim, 1)
ax0.plot(x, x, "r--", lw=1)
ax1.plot(x, x, "r--", lw=1)

ax0.plot(apm_com_tab["pmra_1"], apm_com_tab["pmra_2"], "b.", ms=2)
ax1.plot(apm_com_tab["pmdec_1"], apm_com_tab["pmdec_2"], "b.", ms=2)

ax0.set_xlabel("$\\mu_{\\alpha,*}$ (LOC, $\\mathrm{\\mu as/yr^{-1}}$})", fontsize=15)
ax0.set_ylabel("$\\mu_{\\alpha,*}$ (GLO, $\\mathrm{\\mu as/yr^{-1}}$})", fontsize=15)

ax1.set_xlabel("$\\mu_{\\delta}$ (LOC, $\\mathrm{\\mu as/yr^{-1}}$})", fontsize=15)
ax1.set_ylabel("$\\mu_{\\delta}$ (GLO, $\\mathrm{\\mu as/yr^{-1}}$})", fontsize=15)

plt.tight_layout()
```
